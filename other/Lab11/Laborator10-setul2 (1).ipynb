# Inmplentare regresie

```python
x_text, y_train=data[0].values, data[1].values
```

```python
x_text
```

```python
y_train
```

```python
vect = CountVectorizer(min_df=5, stop_words='english')
x1= vect.fit_transform(x_text)
x2=vect.transform(x_text)
```

```python
print('Reprezentarea ca vectori:\n', x2.toarray())
```

```python
print(f'Dimensiune vocabular: {len(vect.vocabulary_)}')
```

```python
feature_names = vect.get_feature_names()
print('Dimensiunea vocabularului:', len(feature_names))
```

```python
print(feature_names[:20])
```

```python
param_grid = {'C':[0.001, 0.01, 0.1, 1, 10]}
grid = GridSearchCV(LogisticRegression(solver='lbfgs', max_iter=1000), param_grid=param_grid, cv=5, n_jobs=4)
grid.fit(x1, y_train)
print('best cross validation score:', grid.best_score_)
print('best params:', grid.best_params_)
```

# KNeighborsClassifier

```python
parameter_grid = {'n_neighbors': list(range(1,10)), 'p': [1, 2]}
grid_search = GridSearchCV(estimator =KNeighborsClassifier(), param_grid=parameter_grid,scoring='f1', cv=3, iid=False, verbose=10, n_jobs=-1)
cv_results = cross_validate(grid_search, x1, y_train,
                            scoring=('f1','accuracy'),cv=5, return_train_score=True)
```

```python
df=pd.DataFrame()
```

```python
b=cv_results['train_f1'].mean()
c=cv_results['train_accuracy'].mean()
d=cv_results['test_f1'].mean()
e=cv_results['test_accuracy'].mean()
df=df.append( {'Type_of_Search':'GridSearchCv','Model':'KNeighborsClassifier()',
               'train_f1':b, 'train_accuracy':c,
               'test_f1':d,'test_accuracy':e,}, ignore_index=True)
df
```

```python
gs_random = RandomizedSearchCV(estimator=KNeighborsClassifier(), param_distributions=parameter_grid,
                               scoring='f1', cv= 3,n_iter=15, iid=False, verbose=10, n_jobs=-1)
cv_results = cross_validate(gs_random, x1, y_train,
                            scoring=('f1','accuracy'),cv=5, return_train_score=True)
```

```python
b=cv_results['train_f1'].mean()
c=cv_results['train_accuracy'].mean()
d=cv_results['test_f1'].mean()
e=cv_results['test_accuracy'].mean()
df=df.append( {'Type_of_Search':'RandomizedSearchCV','Model':'KNeighborsClassifier()',
               'train_f1':b, 'train_accuracy':c,
               'test_f1':d,'test_accuracy':e}, ignore_index=True)
df
```

# LogisticRegression

```python
param_grid = {'C':[0.001, 0.01, 0.1, 1, 10]}
grid = GridSearchCV(LogisticRegression(solver='lbfgs', max_iter=1000), param_grid=param_grid, cv=5, n_jobs=4, scoring='f1')
cv_results = cross_validate(grid, x1, y_train,
                            scoring=('f1','accuracy'),cv=5, return_train_score=True)
```

```python
b=cv_results['train_f1'].mean()
c=cv_results['train_accuracy'].mean()
d=cv_results['test_f1'].mean()
e=cv_results['test_accuracy'].mean()
df=df.append( {'Type_of_Search':'GridSearchCV','Model':'LogisticRegression()',
               'train_f1':b, 'train_accuracy':c,
               'test_f1':d,'test_accuracy':e}, ignore_index=True)
df
```

```python
gs_random = RandomizedSearchCV(estimator=LogisticRegression(solver='lbfgs', max_iter=1000), param_distributions=param_grid,
                               scoring='f1', cv= 3,n_iter=4, iid=False)
cv_results = cross_validate(gs_random, x1, y_train,
                            scoring=('f1','accuracy'),cv=5, return_train_score=True)
```

```python
b=cv_results['train_f1'].mean()
c=cv_results['train_accuracy'].mean()
d=cv_results['test_f1'].mean()
e=cv_results['test_accuracy'].mean()
df=df.append( {'Type_of_Search':'RandomizedSearchCV','Model':'LogisticRegression()',
               'train_f1':b, 'train_accuracy':c,
               'test_f1':d,'test_accuracy':e}, ignore_index=True)
df
```

# DecisionTreeClassifier

```python
params = {'max_features': ['auto', 'sqrt', 'log2'],
          'min_samples_split':list(range(2,15)), 
          'min_samples_leaf':list(range(2,11)),
         }
grid = GridSearchCV(DecisionTreeClassifier(), param_grid=params, cv=5, n_jobs=4, scoring='f1', verbose=10)
cv_results = cross_validate(grid, x1, y_train,
                            scoring=('f1','accuracy'),cv=5, return_train_score=True)
```

```python
b=cv_results['train_f1'].mean()
c=cv_results['train_accuracy'].mean()
d=cv_results['test_f1'].mean()
e=cv_results['test_accuracy'].mean()
df=df.append( {'Type_of_Search':'GridSearchCV','Model':'DecisionTreeClassifier()',
               'train_f1':b, 'train_accuracy':c,
               'test_f1':d,'test_accuracy':e}, ignore_index=True)
df
```

```python
gs_random = RandomizedSearchCV(estimator=DecisionTreeClassifier(), param_distributions=params,
                               scoring='f1', cv= 3,n_iter=250, iid=False, verbose=10)
cv_results = cross_validate(gs_random, x1, y_train,
                            scoring=('f1','accuracy'),cv=5, return_train_score=True)
```

```python
b=cv_results['train_f1'].mean()
c=cv_results['train_accuracy'].mean()
d=cv_results['test_f1'].mean()
e=cv_results['test_accuracy'].mean()
df=df.append( {'Type_of_Search':'RandomizedSearchCV','Model':'DecisionTreeClassifier()',
               'train_f1':b, 'train_accuracy':c,
               'test_f1':d,'test_accuracy':e}, ignore_index=True)
df
```

# MLPClassifier

```python
paramsM = {'activation' : ['identity', 'logistic', 'tanh', 'relu'], 
          'solver' : ['lbfgs', 'sgd', 'adam'],
          'alpha':[0.1, 0.001, 0.0001]}
grid = GridSearchCV(MLPClassifier(max_iter=10000), param_grid=paramsM, cv=3, n_jobs=-1, scoring='f1', verbose=10)
cv_results = cross_validate(grid, x1, y_train,
                            scoring=('f1','accuracy'),cv=5, return_train_score=True)
```

```python
b=cv_results['train_f1'].mean()
c=cv_results['train_accuracy'].mean()
d=cv_results['test_f1'].mean()
e=cv_results['test_accuracy'].mean()
df=df.append( {'Type_of_Search':'GridSearchCV','Model':'MLPClassifier()',
               'train_f1':b, 'train_accuracy':c,
               'test_f1':d,'test_accuracy':e}, ignore_index=True)
df
```

```python
gs_random = RandomizedSearchCV(estimator=MLPClassifier(max_iter=10000), param_distributions=paramsM,
                               scoring='f1', cv= 3,n_iter=25, iid=False, verbose=10, n_jobs=-1)
cv_results = cross_validate(gs_random, x1, y_train,
                            scoring=('f1','accuracy'),cv=5, return_train_score=True)
```

```python
b=cv_results['train_f1'].mean()
c=cv_results['train_accuracy'].mean()
d=cv_results['test_f1'].mean()
e=cv_results['test_accuracy'].mean()
df=df.append( {'Type_of_Search':'RandomizedSearchCV','Model':'MLPClassifier()',
               'train_f1':b, 'train_accuracy':c,
               'test_f1':d,'test_accuracy':e}, ignore_index=True)
df
```

```python
df['train_f1']=df['train_f1'].abs()
df['train_accuracy']=df['train_accuracy'].abs()
df['test_f1']=df['test_f1'].abs()
df['test_accuracy']=df['test_accuracy'].abs()
df
```
