```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn import linear_model
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import cross_validate
import sklearn
```

```python
ad1="http://archive.ics.uci.edu/ml/machine-learning-databases/communities/communities.data"
#names=['col']*128
df2=pd.read_csv(ad1, header=None)
df2[:5]
```

```python
df2.info()
```

```python
df2=df2.drop(df2.columns[[0, 1, 2, 3, 4]], 1)
df2
```

```python
df2=df2.replace('?', value=np.nan)
df2.head()
```

```python
df2 = df2.dropna(axis='columns')
df2[0:5]
```

```python
mat=df2.values
X=mat[:, 0:-1]
y=mat[:, -1]
scaler = MinMaxScaler()
scaler.fit(X)
X = scaler.transform(X)
```

# LinearRegression

```python
parameter_grid = {'fit_intercept': [True, False], 'normalize': [True, False], 'copy_X': [True, False]}
grid_search = GridSearchCV(estimator = LinearRegression(), param_grid=parameter_grid, cv=3, iid=False)
```

```python
cv_results = cross_validate(grid_search, X, y,
                            scoring=('neg_mean_squared_error', 'neg_mean_absolute_error','neg_median_absolute_error'),
                            cv=5, return_train_score=True)
a=cv_results['train_neg_mean_squared_error'].mean()
b=cv_results['train_neg_mean_absolute_error'].mean()
c=cv_results['train_neg_median_absolute_error'].mean()
d=cv_results['test_neg_mean_squared_error'].mean()
e=cv_results['test_neg_mean_absolute_error'].mean()
f=cv_results['test_neg_median_absolute_error'].mean()
g=cv_results['fit_time'].mean()
```

```python
df=pd.DataFrame()
df=df.append( {'Type_of_Search':'GridSearchCv','Model':'LinearRegression()','train_neg_mean_squared_error':a,
               'train_neg_mean_absolute_error':b, 'train_neg_median_absolute_error':c,
               'test_neg_mean_squared_error':d,'test_neg_mean_absolute_error':e,
               'test_neg_median_absolute_error':f, 'fit_time':g}, ignore_index=True)
df
```

```python
gs_random = RandomizedSearchCV(estimator=LinearRegression(), param_distributions=parameter_grid, cv= 3,n_iter=5, iid=False)
cv_random = cross_validate(gs_random, X, y,
                            scoring=('neg_mean_squared_error', 'neg_mean_absolute_error','neg_median_absolute_error'),
                            cv=5, return_train_score=True)
a=cv_random['train_neg_mean_squared_error'].mean()
b=cv_random['train_neg_mean_absolute_error'].mean()
c=cv_random['train_neg_median_absolute_error'].mean()
d=cv_random['test_neg_mean_squared_error'].mean()
e=cv_random['test_neg_mean_absolute_error'].mean()
f=cv_random['test_neg_median_absolute_error'].mean()
g=cv_random['fit_time'].mean()
df=df.append( {'Type_of_Search':'RandomizedSearchCV','Model':'LinearRegression()','train_neg_mean_squared_error':a,
               'train_neg_mean_absolute_error':b, 'train_neg_median_absolute_error':c,
               'test_neg_mean_squared_error':d,'test_neg_mean_absolute_error':e,
               'test_neg_median_absolute_error':f, 'fit_time':g}, ignore_index=True)
df
```

```python
def highlight_max(df):
    is_max = df == df.max()
    return ['background-color: red' if v else '' for v in is_max]

def highlight_min(df):
    is_min = df == df.min()
    return ['background-color: yellow ' if a else '' for a in is_min]


df.style.apply(highlight_max)\
.apply(highlight_min)
```
