# 1) Regresia liniară

Regresia liniară este utilizată pentru a găsi relația liniară între o valoare target și unul sau mai mulți predictori. Există două tipuri de regresie liniară: simpla și multipla.

Regresia liniară simplă este utilă pentru a găsi relația între două variabile continue. Unul este predictor sau variabilă independentă, iar alta este variabila răspuns sau dependentă.Relația dintre două variabile se spune că este deterministă dacă o variabilă poate fi exprimată cu exactitate de cealaltă. De exemplu, folosind temperatura în grade Celsius, este posibil să se prezica cu precizie. Relația statistică nu este exactă în determinarea relației dintre două variabile. De exemplu, relația dintre înălțime și greutate.

Ideea principală este obținerea unei linii care se potrivește cel mai bine datelor. Cea mai bună linie de potrivire este cea pentru care eroarea de predicție totală (toate punctele de date) este cât mai mică posibil. Eroare este distanța dintre punctul și linia de regresie.


Termenul de regresie multipla se foloseste in cazul in care avem doi sau mai multi predictori si un singur raspuns.Există, de asemenea, modele de regresie cu două sau mai multe variabile de răspuns. Aceste modele sunt denumite de obicei modele de regresie multivariate. 

Acest tip de regresie se gaseste in biblioteca sklearn.linear_model sub denumirea de LinearRegression. 

sklearn.linear_model.LinearRegression(fit_intercept=True, normalize=False, copy_X=True, n_jobs=None)

Parametrii:

-fit_intercept-dacă se calculează interceptul pentru acest model. Dacă este setat la False, interceptarea nu va fi utilizată în calcule (de exemplu, datele vor fi deja centrate), valoare booleana, default va avea valoare True.

-normalize-Acest parametru este ignorat atunci când fit_intercept este setat la False. Dacă este adevărat, regresorii X vor fi normalizați înainte de regresie prin scăderea mediei și împărțirea după norma l2. Valoare booleana, default va avea valoare False. 

-copy_X-Dacă este adevărat, X va fi copiat; altfel, va fi suprascris. Valoare booleana, default va avea valoare True. 

-n_jobs-numarul de job-uri pe care procesorul le va face, valoarea -1 inseamna folosirea tuturor resurselor. Acesta are o valoare intreaga sau chiar None.

# 2) KNeighborsRegressor

Algoritmul k-NN este o metodă neparametrizata utilizată pentru clasificare și regresie. În ambele cazuri, intrarea constă în exemplele de instruire k cele mai apropiate din spațiul caracteristic. Producția depinde de faptul dacă k-NN este folosit pentru clasificare sau regresie: in regresia k-NN, ieșirea este valoarea proprietății pentru obiect. Această valoare reprezintă media valorilor celor mai apropiați vecini. k-NN este un tip de învățare bazată pe instanțe sau învățare leneșă, în care funcția este aproximată doar la nivel local și toate calculele sunt amânate până la clasificare. Algoritmul k-NN este printre cei mai simpli dintre toți algoritmi de învățare a mașinilor.

In regresia k-NN, algoritmul k-NN este utilizat pentru estimarea variabilelor continue. Un astfel de algoritm utilizează o medie ponderată a celor mai apropiați vecini k, ponderată de inversul distanței lor. Acest algoritm funcționează după cum urmează:

- se calculeaza distanța Euclidiană sau Mahalanobi de la exemplul de interogare la exemplele etichetate.

- se ordoneaza etichetele prin creșterea distanței.

- se gaseste numarul optim euristic al celor mai apropiați vecini, bazat pe RMSE. Acest lucru se face folosind validarea încrucișată.

Acest tip de regresie se gaseste in biblioteca sklearn.neighbors sub denumirea de KNeighborsRegressor.

sklearn.neighbors.KNeighborsRegressor(n_neighbors=5, weights='uniform', algorithm='auto', leaf_size=30, warn_on_equidistant=True)

Parametrii:
- n_neighbors : int, optional (default = 5): Numărul de vecini care vor fi utilizați în mod implicit pentru interogările k_neighbors.
- weights: string (default='uniform'): funcția de greutate utilizată în predicție. Valorile posibile:

-"uniform": greutăți uniforme. Toate punctele din fiecare cartier sunt ponderate în mod egal; 

-"distance": punctele de greutate prin inversul distanței lor. în acest caz, vecinii mai apropiați ai unui punct de interogare vor avea o influență mai mare decât vecinii aflați mai departe.
- algorithm : {'auto', 'ball_tree', 'kd_tree', 'brute'}, optional. Algoritmul folosit pentru a calcula cei mai apropiați vecini:

-"ball_tree" va folosi BallTree; 

-'kd_tree' va folosi KDtree;

-"brute" va folosi o căutare de forțe brute; 

-"auto" va încerca să decidă cel mai potrivit algoritm pe baza valorilor transmise pentru a se potrivi cu metoda.
- leaf_size : int, optional (default = 30). Folosit la BallTree sau KDTree. Acest lucru poate afecta viteza construcției și interogarea, precum și memoria necesară pentru stocare. Valoarea optimă depinde de natura problemei.
- metric : string or DistanceMetric object (default='minkowski'). Metrica de distanță care trebuie utilizată pentru arbore. Valoarea implicită este minkowski, iar cu p = 2 este echivalentă cu ecuația standard Euclidiană.
- p : int, optional (default = 2). Atunci când p = 1, aceasta este echivalentă cu utilizarea manhattan_distance (l1) și a duratei euclidiene (l2) pentru p = 2. Pentru p arbitrar, se utilizează minkowski_distance (l_p).
- metric_params : dict, optional (default = None).  Cuvinte cheie suplimentare pentru funcția de metrică.

# 3) DecisionTreeRegressor

Decision Trees (DTs) sunt o metodă de învățare neparametrică supravegheată folosită pentru clasificare și regresie. Scopul este de a crea un model care prezice valoarea unei variabile țintă prin învățarea unor simple reguli de decizie deduse din caracteristicile de date.
De asemenea, arborii de decizie pot fi aplicați și pentru problemele de regresie, utilizând clasa DecisionTreeRegressor.
Ca și în setarea de clasificare, metoda de potrivire va lua ca arhitecturi de argument X și y, numai că în acest caz y este de așteptat să aibă valori în virgulă mobilă în loc de valori întregi.
Putem observa că dacă adâncimea maximă a copacului (controlată de parametrul max_depth) este setată prea înaltă, arborii de decizie învață detalii prea fine ale datelor de antrenament și învață din noise, adică se suprapun.


# Parameters

-> criterion : string, optional (default="mse")

Funcția de măsurare a calității unei divizări. Singurul criteriu acceptat este "mse" pentru eroarea medie pătrată, care este egală cu reducerea varianței ca criteriu de selecție a caracteristicilor.

-> splitter : string, optional (default="best")

Strategia folosită pentru alegerea divizării la fiecare nod. Strategiile acceptate sunt "best" pentru a alege cea mai bună divizare și "random" pentru a alege cea mai bună diviziune aleatorie.

-> max_features : int, float, string or None, optional (default=None)

Numărul de caracteristici care trebuie luate în considerare atunci când căutați cea mai bună împărțire:

Dacă "int", atunci luați în considerare caracteristicile max_features la fiecare split.

Dacă "float", atunci max_features este un procent și caracteristicile int (max_features * n_features) sunt luate în considerare la fiecare split.

Dacă "auto", atunci max_features = n_features.

Dacă "sqrt", atunci max_features = sqrt (n_features).

Dacă "log2", atunci max_features = log2 (n_features).

Dacă nu există, atunci max_features = n_features.

-> max_depth : int or None, optional (default=None)

Adâncimea maximă a arborelui. Dacă nu există, atunci nodurile sunt extinse până când toate frunzele sunt pure sau până când toate frunzele conțin mai puțin decât probele min_samples_split. Ignorat dacă max_leaf_nodes nu este None.

-> min_samples_split : int, optional (default=2)

Numărul minim de eșantioane necesare pentru divizarea unui nod intern.

-> min_samples_leaf : int, optional (default=1)

Numărul minim de eșantioane necesare la un nod frunza.

-> min_weight_fraction_leaf : float, optional (default=0.)

Fracțiunea minimă ponderată a probelor de intrare care trebuie să fie la un nod de frunze.

-> max_leaf_nodes : int or None, optional (default=None)

Dacă "none" există un număr nelimitat de noduri frunze. Dacă nu este "none" atunci max_depth va fi ignorat.
    

-> random_state : int, RandomState instance or None, optional (default=None)

Dacă este "int", random_state este generatorul de numere aleatoare; Dacă este instanța "RandomState", random_state este generatorul de numere aleatoare; Dacă nu există, generatorul de numere aleatoare este instanța RandomState folosită de np.random

-> presort : bool, optional (default=False)

Pentru setările implicite ale unui arbore de decizie pe seturi de date mari, setarea acestuia la "true" poate încetini procesul de antrenament. Atunci când utilizați un set de date mai mic sau o adâncime limitată, acest lucru poate accelera procesul de antrenament.

# 4) SVR (Support Vector Regression)


Metoda Support Vector Classification poate fi extinsă pentru a rezolva problemele de regresie. Această metodă se numește Support Vector Regression.

Modelul produs de SVC depinde doar de un subset al datelor de antrenament, deoarece funcția de cost pentru construirea modelului nu este interesata de punctele de antrenare care se află dincolo de margine. În mod analog, modelul produs de SVR depinde numai de un subset de date de antrenament, deoarece funcția de cost pentru construirea modelului ignoră toate datele de antrenament apropiate de predicția modelului.

Există trei implementări diferite ale SVR: SVR, NuSVR și LinearSVR. LinearSVR oferă o implementare mai rapidă decât SVR, dar are în vedere doar nucleele liniare, în timp ce NuSVR implementează o formulă puțin diferită decât SVR și LinearSVR.

# Parameters

-> C : float, optional (default=1.0)

Parametrul de penalizare C al termenului de eroare.

-> epsilon : float, optional (default=0.1)

Specifică epsilonul în care nu este asociată nici o penalitate în funcția de pierdere a setului de antrenare cu punctele prezise la o distanță epsilon față de valoarea reală.

-> kernel : string, optional (default='rbf')

Trebuie să fie una dintre cele "linear", "poli", "rbf", "sigmoid", "precomputed" sau "callable". Dacă nu se dă niciunul, se va folosi "rbf". Dacă este dat un apel callable, acesta este folosit pentru a precompune matricea kernel-ului.

-> degree : int, optional (default=3)

Gradul funcției kernelului polinomial ("poli"). Ignorat de toate celelalte kernel-uri.

-> gamma : float, optional (default='auto')

Coeficientul kernel pentru "rbf", "poly" și "sigmoid". Dacă gama este "auto", atunci vor fi utilizate 1 / n_features.

-> coef0 : float, optional (default=0.0)

Termen independent în funcția de kernel. Este semnificativă numai în "poli" și "sigmoid".

-> shrinking : boolean, optional (default=True)

-> tol : float, optional (default=1e-3)

-> cache_size : float, optional

Dimensiunea memoriei cache a kernel (în MB).

-> verbose : bool, default: False

-> max_iter : int, optional (default=-1)

Limita maximă a iterațiilor sau -1 pentru nici o limită.

# 5) MLPRegressor

Un perceptron multistrat (MLP) este o clasă a rețelei neuronale artificiale. Un MLP constă din cel puțin trei straturi de noduri: un strat de intrare, un strat ascuns și un strat de ieșire. Cu excepția nodurilor de intrare, fiecare nod este un neuron care utilizează o funcție de activare neliniară. MLP utilizează o tehnică de învățare supervizata numită backpropagation pentru formare.  Straturile sale multiple și activarea neliniară disting MLP de un perceptron liniar. Poate distinge date care nu pot fi separate liniar.

Un perceptron multistrat este o rețea neurală profundă, artificială. Este compus din mai mult de un perceptron. Acestea sunt compuse dintr-un strat de intrare pentru a recepționa semnalul, un strat de ieșire care face o decizie sau o predicție despre intrare, iar între cele două, un număr arbitrar de straturi ascunse, care sunt adevăratul motor computational al MLP-ului. MLP-urile cu un strat ascuns sunt capabile să aproximeze orice funcție continuă.

Folosind backpropagation și chain rule, derivatele parțiale ale funcției de eroare, diferitele greutăți și prejudecăți sunt propagate înapoi prin MLP. Acest act de diferențiere ne oferă un gradient sau un peisaj de eroare, de-a lungul căruia parametrii pot fi ajustați în timp ce aceștia deplasează MLP-ul cu un pas mai aproape de minimul de eroare. Acest lucru se poate face cu orice algoritm de optimizare bazat pe gradienți, cum ar fi coborârea stochastică a gradientului. Această stare este cunoscută drept convergență.

Parametrii:

- hidden_layer_sizes : tuple, length = n_layers - 2, default (100,).Elementul i reprezintă numărul de neuroni din stratul i.

- activation : {‘identity’, ‘logistic’, ‘tanh’, ‘relu’}, default ‘relu’

Funcția de activare pentru stratul ascuns.

"identity", activare no-op, returnează f (x) = x
"logistic", funcția sigmoid logistică, returnează f (x) = 1 / (1 + exp (-x)).
"tanh", funcția tan hiperbolică, returnează f (x) = tanh (x).
"relu", funcția liniară rectificată, returnează f (x) = max (0, x)

- solver : {‘lbfgs’, ‘sgd’, ‘adam’}, default ‘adam’

Solver pentru optimizarea greutății.

"Lbfgs" este un optimizator în familia metodelor cvasi-Newton.
'Sgd' se referă la coborârea stochastică a gradientului.
"Adam" se referă la un optimizator bazat pe gradient stochastic propus de Kingma, Diederik și Jimmy Ba

- alpha : float, optional, default 0.0001

Valoarea penalizării L2 (termenul de regularizare).

- batch_size : int, optional, default ‘auto’

Dimensiunea loturilor mini pentru optimizatori stochastici. Dacă solverul este "lbfgs", clasificatorul nu va folosi mini lot. Când este setat la "auto", batch_size = min (200, n_samples)

- learning_rate : {‘constant’, ‘invscaling’, ‘adaptive’}, default ‘constant’

Rata de învățare pentru actualizări în greutate.

'Constant' este o rată constantă de învățare dată de 'learning_rate_init'.

‘invscaling’ scade treptat rata de învățare learning_rate_ la fiecare pas de timp 't' folosind un exponent invers de scalare al 'power_t'. effective_learning_rate = learning_rate_init / pow (t, power_t)

‘adaptive’ menține rata de învățare constantă la "learning_rate_init" atâta timp cât pierderea de formare continuă să scadă. 

- power_t : double, optional, default 0.5

Exponentul pentru rata de invatare scalare inversa. Se folosește la actualizarea ratei efective de învățare când learning_rate este setat la 'int scaling'. Se folosește numai atunci când solver = 'sgd'.

- max_iter : int, optional, default 200

Numărul maxim de iterații.Pentru solverii stochastice ('sgd', 'adam'), acest lucru determină numărul de de câte ori va fi folosit fiecare punct de date, nu numărul de pași de gradient.

- random_state : int, RandomState instance or None, optional, default None

Dacă instanța RandomState, random_state este generatorul de numere aleatoare; Dacă nu există, generatorul de numere aleatoare este instanța RandomState folosită de np.random.
