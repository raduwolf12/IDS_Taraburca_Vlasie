```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn import linear_model
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import cross_validate
import sklearn
```

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/cpu-performance/machine.data', sep=",", header=None)
print("Describe Dataset","\n")
print(data.info())
print(data.head())
print(data.describe())
```

```python
data[:5]
```

```python
data=data.drop(columns=[0, 1, 9])
```

```python
data[:5]
```

```python
mat=data.values

X=mat[:, 0:-1]
scaler = MinMaxScaler()
scaler.fit(X)
X = scaler.transform(X)
print()
print(X.shape)

y=mat[:, -1]
print(y.shape)
```

# 1)Linear Regression

```python
parameter_grid = {'fit_intercept': [True, False],
                  'normalize': [True, False],
                  'copy_X': [True, False]}
```

```python
grid_search = GridSearchCV(estimator = LinearRegression(), param_grid=parameter_grid, cv=3, iid=False)
```

```python
cv_results = cross_validate(grid_search, X, y,
                            scoring=('neg_mean_squared_error', 'neg_mean_absolute_error','neg_median_absolute_error'),
                            cv=5, return_train_score=True)

a=cv_results['train_neg_mean_squared_error'].mean()
b=cv_results['train_neg_mean_absolute_error'].mean()
c=cv_results['train_neg_median_absolute_error'].mean()
d=cv_results['test_neg_mean_squared_error'].mean()
e=cv_results['test_neg_mean_absolute_error'].mean()
f=cv_results['test_neg_median_absolute_error'].mean()
g=cv_results['fit_time'].mean()
```

```python
df=pd.DataFrame()
df=df.append( {'Type_of_Search':'GridSearchCv',
               'Model':'LinearRegression',
               'train_neg_mean_squared_error':a,
               'train_neg_mean_absolute_error':b, 
               'train_neg_median_absolute_error':c,
               'test_neg_mean_squared_error':d,
               'test_neg_mean_absolute_error':e,
               'test_neg_median_absolute_error':f,
               'fit_time':g}, ignore_index=True)
df
```

```python
gs_random = RandomizedSearchCV(estimator=LinearRegression(), param_distributions=parameter_grid, cv= 3,n_iter=5, iid=False)
```

```python
cv_random = cross_validate(gs_random, X, y,
                            scoring=('neg_mean_squared_error', 'neg_mean_absolute_error','neg_median_absolute_error'),
                            cv=5, return_train_score=True)


a=cv_random['train_neg_mean_squared_error'].mean()
b=cv_random['train_neg_mean_absolute_error'].mean()
c=cv_random['train_neg_median_absolute_error'].mean()
d=cv_random['test_neg_mean_squared_error'].mean()
e=cv_random['test_neg_mean_absolute_error'].mean()
f=cv_random['test_neg_median_absolute_error'].mean()
g=cv_random['fit_time'].mean()


df=df.append( {'Type_of_Search':'RandomizedSearchCV',
               'Model':'LinearRegression',
               'train_neg_mean_squared_error':a,
               'train_neg_mean_absolute_error':b,
               'train_neg_median_absolute_error':c,
               'test_neg_mean_squared_error':d,
               'test_neg_mean_absolute_error':e,
               'test_neg_median_absolute_error':f, 
               'fit_time':g}, ignore_index=True)
df
```

```python
def highlight_max(df):
    is_max = df == df.max()
    return ['background-color: red' if v else '' for v in is_max]
```

```python
def highlight_min(df):
    is_min = df == df.min()
    return ['background-color: lime ' if a else '' for a in is_min]
```

```python

df.style.apply(highlight_max, subset=['fit_time','test_mean_absolute_error','test_mean_squared_error','test_median_absolute_error','train_mean_absolute_error','train_mean_squared_error','train_median_absolute_error'])\
.apply(highlight_min,subset=['fit_time','test_mean_absolute_error','test_mean_squared_error','test_median_absolute_error','train_mean_absolute_error','train_mean_squared_error','train_median_absolute_error'] )\
.hide_index()
```
